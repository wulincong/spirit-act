```python
import h5py
import glob
import os

# 替换为你实际的数据集路径
dataset_dir = 'dataset/sim_transfer_cube_scripted' 

# 找到第一个 hdf5 文件
files = glob.glob(os.path.join(dataset_dir, '*.hdf5'))
if not files:
    print("未找到 hdf5 文件，请检查路径")
else:
    file_path = files[0]
    print(f"正在检查文件: {file_path}")
    
    with h5py.File(file_path, 'r') as f:
        # 查看 observations/images 下有哪些 key
        if 'observations/images' in f:
            cam_names = list(f['observations/images'].keys())
            print(f"\n✅ 你的真实相机名称列表: {cam_names}")
        else:
            print("\n❌ 未找到 /observations/images 路径")
```
